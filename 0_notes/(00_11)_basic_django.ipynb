# ðŸ What is Django 5?

**Django** is a **high-level web framework** written in **Python** that encourages **rapid development** and **clean, pragmatic design**.

Django emphasizes **reusability** and the **"Don't Repeat Yourself" (DRY)** principle, which means it tries to keep our code **simple** and **non-repetitive**.

It follows the **Model-View-Template (MVT)** architectural pattern.

## Video-4: Creating Django Project
---
---

**Django Project:**
- A Django  Project may contain multiple project Application, which means a group of Application and files is called as Django Project.
- An Application is a part of Django Project

**Example: School Management System**
- Registration App
- Fees App
- Exam App
- Result App

**Create Django Project**
1. If No project folder exists:
``` django-admin startproject projectname ```

2. If project folder exists:
``` django-admin startproject projectname . ```

### Only in this tutorial
**First time Project Setup:**
- Install python 
- Create django-5 project folder
- Install virtual env: > ``` pip install virtualenv ```
- Create virtual env in django-5 folder: > ``` virtualenv venv ```
- Activate virtual env: > ``` venv\Scripts\activate ```
- Install django: > ``` pip install django ```
- Create .gitignore file
- Create requirements.txt file: > ``` pip freeze > requirements.txt ```

**Every time Project Setup:**
- Open django-5 folder
- Create a new project folder: > ``` mkdir projectname & cd projectname ```
- Create the project: > ``` django-admin startproject projectname . ```
- Create Application: > ``` python manage.py startapp appname ```
- Run Server: > ``` python manage.py runserver ```
- Create Migration files: > ``` python manage.py makemigrations ```
- Make Migrations: > ``` python manage.py migrate ```
- Create Super user: > ``` python manage.py createsuperuser ```

## Video-5 (ch4): Project Directory Structure in Detail
---
---

**Project file and folder details**

- `Outer ch4` is the Outer Project folder or Root Directory
- `Inner ch4` is the Inner Project folder
- `__init__.py` file make inner ch4 folder as python package
- `wsgi.py` (Web Server Gateway Interface) file work as interface between this project and the web server. How this web application communicate with the web server, is written in this file. The standard for python synchronous operation is provided in this file.
- `asgi.py` (Asynchronous Server Gateway Interface) file is the successor of wsgi.py file. The standard for python synchronous and asynchronous both application or operation is provided in this file. work as interface between the application and the web server. How this web application communicate with the web server, is written in this file.
- `settings.py` file includes all the setting of the web application
- `urls.py` file is used to create url
- `manage.py` file is a command line utility

## Video-11 (ch8 & ch9): URL Patterns in Django

**Passing Keyword argument with urls:**

ðŸ“‚ch8/myapp/views.py

---
```python

from django.shortcuts import render
from django.http import HttpResponse

def learn_django(request):
    return HttpResponse('<h1>Hello Django</h1>')

def learn_django2(request, **kwargs):
    status = kwargs.get('status')
    return HttpResponse(f'<h1>Hello Django {status}</h1>')

```

ðŸ“‚ch8/urls.py

---
```python

from django.contrib import admin
from django.urls import path
from app1.views import learn_django, learn_django2

urlpatterns = [
    path('admin/', admin.site.urls),
    path('py1/',learn_django),
    path('py2/', learn_django),
    path('dj/', learn_django2, {'status': 'OK'})
]

```
