<h2>What is Django?</h2>
<p>Django is a <strong>high-level Python web framework</strong> that allows developers to build secure, scalable, and maintainable web applications quickly.</p>
<blockquote>‚ÄúDon‚Äôt Repeat Yourself (DRY)‚Äù</blockquote>

<hr>

<h2>Why use Django?</h2>
<p>Django comes with many features built-in, so you don‚Äôt have to create them from scratch.</p>

<h3>Key Advantages</h3>
<ul>
    <li>Fast development</li>
    <li>Secure by default</li>
    <li>Scalable</li>
    <li>Clean and structured project layout</li>
    <li>Large community and documentation</li>
</ul>

<hr>

<h2>What can you build with Django?</h2>
<ul>
    <li>Blogs and content websites</li>
    <li>Authentication systems (login / signup)</li>
    <li>Dashboards and admin panels</li>
    <li>REST APIs</li>
    <li>E-commerce platforms</li>
    <li>Backend for mobile apps</li>
</ul>

<h3>Popular Platforms Using Django</h3>
<ul>
    <li>Instagram</li>
    <li>Pinterest</li>
    <li>Mozilla</li>
</ul>

<hr>

<h2>Django Architecture (MVT)</h2>
<p>Django follows the <strong>MVT (Model-View-Template)</strong> pattern:</p>

<table>
    <tr>
        <th>Component</th>
        <th>Meaning</th>
        <th>Responsibility</th>
    </tr>
    <tr>
        <td>Model</td>
        <td>Data Layer</td>
        <td>Database schema and queries</td>
    </tr>
    <tr>
        <td>View</td>
        <td>Logic Layer</td>
        <td>Handles request & response</td>
    </tr>
    <tr>
        <td>Template</td>
        <td>Presentation Layer</td>
        <td>HTML UI</td>
    </tr>
</table>

<hr>

<h2>How Django Works (Request Flow)</h2>
<ol>
    <li>User sends a request (URL)</li>
    <li>Django routes the URL via <code>urls.py</code></li>
    <li>A view function/class is executed</li>
    <li>View interacts with models (database)</li>
    <li>Response (HTML/JSON) is returned</li>
</ol>

<h2>1Ô∏è‚É£ Check Python & Django</h2>
<pre><code class="command">python --version           # Check Python version
django-admin --version     # Check Django version</code></pre>
<div class="note">Install Django if not present: <code>pip install django</code></div>

<hr>

<h2>2Ô∏è‚É£ Project Setup</h2>
<pre><code class="command"># Create project folder
mkdir project_folder
cd project_folder

# Create & activate virtual environment
python -m venv venv
venv\Scripts\activate

# Install Django
pip install django

# Start Django project
django-admin startproject project_folder .

# Run server
python manage.py runserver
python manage.py runserver 4000  # custom port</code></pre>
<div class="note">Outer folder = root directory, Inner folder = Django project.</div>

<hr>

<h2>3Ô∏è‚É£ App Creation</h2>
<pre><code class="command">python manage.py startapp app_name</code></pre>
<div class="note">Add app to <code>INSTALLED_APPS</code> in settings.py:</div>
<pre><code>INSTALLED_APPS = [
    ...
    'app_name',
]</code></pre>

<hr>

<h2>4Ô∏è‚É£ Database Commands</h2>
<pre><code class="command">python manage.py makemigrations      # create migrations for models
python manage.py migrate                 # apply migrations to DB
python manage.py sqlmigrate app_name 0001  # see SQL of migration
python manage.py showmigrations           # list applied migrations</code></pre>

<hr>

<h2>5Ô∏è‚É£ Superuser & Admin</h2>
<pre><code class="command">python manage.py createsuperuser       # create admin user
python manage.py runserver
# Access admin panel at http://127.0.0.1:8000/admin/</code></pre>

<hr>

<h2>6Ô∏è‚É£ Virtual Environment</h2>
<pre><code class="command"># Activate
venv\Scripts\activate

# Deactivate
deactivate

# Check installed packages
pip freeze

# Save packages
pip freeze &gt; requirements.txt

# Install from requirements
pip install -r requirements.txt</code></pre>

<hr>

<h2>7Ô∏è‚É£ Common Useful Commands</h2>
<ul>
    <li><span class="command">python manage.py shell</span> ‚Äî open Django interactive shell</li>
    <li><span class="command">python manage.py check</span> ‚Äî check project for errors</li>
    <li><span class="command">python manage.py runserver</span> ‚Äî run development server</li>
    <li><span class="command">python manage.py collectstatic</span> ‚Äî collect static files for production</li>
    <li><span class="command">python manage.py test</span> ‚Äî run tests</li>
</ul>

<h2>8Ô∏è‚É£ Important Django Files</h2>
<ul>
    <li><strong>manage.py</strong> ‚Äî command-line utility</li>
    <li><strong>settings.py</strong> ‚Äî project configuration (DB, apps, templates, static files)</li>
    <li><strong>urls.py</strong> ‚Äî URL routing</li>
    <li><strong>wsgi.py</strong> ‚Äî production server interface (sync)</li>
    <li><strong>asgi.py</strong> ‚Äî async-capable server interface</li>
    <li><strong>__init__.py</strong> ‚Äî marks folder as Python package</li>
    <li><strong>apps.py</strong> ‚Äî app configuration</li>
    <li><strong>models.py</strong> ‚Äî database models</li>
    <li><strong>views.py</strong> ‚Äî request handling</li>
    <li><strong>admin.py</strong> ‚Äî register models for admin panel</li>
    <li><strong>migrations/</strong> ‚Äî track database changes</li>
</ul>
